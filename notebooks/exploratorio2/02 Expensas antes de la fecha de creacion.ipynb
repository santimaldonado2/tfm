```python
import pandas as pd
import numpy as np
from plotnine import *

%matplotlib inline
```

```python
data = pd.read_csv("../../data/processed/expensas_full.csv",
                  dtype = {
                                "expensa_id": "object",
                                "unidad_id": "object",
                                "propietario_id": "object",
                                "inquilino_id": "object",
                                "consorcio_id": "object",
                                "consorcio_usa_fondo": "bool",
                                "consorcio_solo_muestra_cat": "bool",
                                "unidad_tipo":"object",
                                "expensa_concepto": "object",
                                "expensa_mes":"object",
                                "expensa_anio":"object",
                                "unidad_propietario":"object",
                                "unidad_inquilino":"object",
                                "expensa_version":"object"
                            })
```

```python
data.dtypes
```

```python
data.consorcio_fecha_creacion = pd.to_datetime(data.consorcio_fecha_creacion, errors="coerce")
```

```python
data["expensa_fecha"] = pd.to_datetime(data.expensa_anio + "-" +data.expensa_mes + '-01')
```

```python
data.expensa_primer_vencimiento = pd.to_datetime(data.expensa_primer_vencimiento)
```

```python
data.dtypes
```

```python
data = data.loc[~data.consorcio_fecha_creacion.isna(), :]
```

```python
supuestas_moras_cargadas = data.loc[data.consorcio_fecha_creacion > data.expensa_primer_vencimiento]
```
