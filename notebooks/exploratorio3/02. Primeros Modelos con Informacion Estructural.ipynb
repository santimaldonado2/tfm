```python
import pandas as pd
import numpy as np
from plotnine import *

%matplotlib inline
```

```python
data = pd.read_csv("../../data/processed/expensas_full.csv",
                  dtype = {
                                "expensa_id": "object",
                                "unidad_id": "object",
                                "propietario_id": "object",
                                "inquilino_id": "object",
                                "consorcio_id": "object",
                                "consorcio_usa_fondo": "bool",
                                "consorcio_solo_muestra_cat": "bool",
                                "unidad_tipo":"object",
                                "expensa_concepto": "object",
                                "expensa_mes":"object",
                                "expensa_anio":"object",
                                "unidad_propietario":"object",
                                "unidad_inquilino":"object",
                                "expensa_version":"object"
                            },
                  parse_dates = ["consorcio_fecha_creacion", "expensa_primer_vencimiento", "expensa_segundo_vencimiento"])
data.pago_fecha = pd.to_datetime(data.pago_fecha, errors="coerce")
data.expensa_mes = data.expensa_mes.str.pad(width=2, side = "left", fillchar= "0")
data["expensa_fecha"] = pd.to_datetime(data.expensa_anio + "-" + data.expensa_mes + "-01")
data["expensa_mes_pago"] = pd.to_datetime(data.expensa_primer_vencimiento.dt.year.map(str) + '-'+ data.expensa_primer_vencimiento.dt.month.map(str) + '-01')
data.loc[data.pago_fecha.isna() | (data.pago_fecha >= (data.expensa_mes_pago + pd.DateOffset(months=1))), "target"] = 1
data.loc[data.target.isna(), "target"] = 0
data[["expensa_mes","expensa_anio", "expensa_mes_pago", "expensa_primer_vencimiento", "expensa_segundo_vencimiento", "pago_fecha", "pago_monto", "target"]]
```

```python
data.columns
```

```python
df = data[['consorcio_id', 'consorcio_nombre', 'unidad_id',
       'unidad_denominacion','unidad_tipo','expensa_id',
       'expensa_proporcion', 'expensa_mes', 'expensa_fecha', 'expensa_mes_pago', 
       'expensa_version', 'expensa_concepto', 'pago_fecha', 'pago_metodo','target']]
```

```python
df["expensa_mes_pago_anterior"] = df.expensa_mes_pago - pd.DateOffset(months=1)
```

```python
df.loc[df.pago_metodo.isna(), "pago_metodo"] = "Impago"
```

```python
estadisticas_consorcio = df[["consorcio_id", "expensa_mes_pago", "target", "expensa_id", "pago_metodo"]]
```

```python
estadisticas_consorcio["pago_metodo_2"] = estadisticas_consorcio.pago_metodo
estadisticas_consorcio.loc[estadisticas_consorcio.target==1, "pago_metodo_2"] = "Impago"
estadisticas_consorcio["efectivo"] = estadisticas_consorcio.pago_metodo_2 == "Efectivo"
```

```python
impagos_por_mes = estadisticas_consorcio[["consorcio_id","expensa_mes_pago","target"]].groupby(["consorcio_id","expensa_mes_pago"]).sum()
total_pagos_por_mes =  estadisticas_consorcio[["consorcio_id","expensa_mes_pago","expensa_id"]].groupby(["consorcio_id","expensa_mes_pago"]).count()
pagos_efectivos_por_mes = estadisticas_consorcio[["consorcio_id","expensa_mes_pago","efectivo"]].groupby(["consorcio_id","expensa_mes_pago"]).sum()
```

```python
resumen_consorcios = pd.merge(impagos_por_mes, total_pagos_por_mes, left_index=True, right_index=True)
resumen_consorcios = pd.merge(resumen_consorcios, pagos_efectivos_por_mes, left_index=True, right_index=True)
resumen_consorcios.columns = ["impagos", "cantidad_expensas", "cantidad_efectivo"]
resumen_consorcios["prop_impagos"] = resumen_consorcios.impagos / resumen_consorcios.cantidad_expensas
resumen_consorcios["prop_efectivo"] = resumen_consorcios.cantidad_efectivo / resumen_consorcios.cantidad_expensas
#resumen_consorcios = resumen_consorcios[["cantidad_expensas","prop_impagos", "prop_efectivo"]]
resumen_consorcios.to_csv("../../data/processed/resumen_consorcios.csv")
```

```python
def get_cum_by_unidad(unidad_id):
    data_unidad = df.loc[data.unidad_id == unidad_id, ["unidad_id","expensa_mes_pago", "target"]].sort_values("expensa_mes_pago")
    data_unidad["cum_impagos"] = data_unidad.target.cumsum()
    data_unidad["total_pagos"] = range(1,data_unidad.shape[0] + 1)
    data_unidad["prop_impagos"] = data_unidad.cum_impagos / data_unidad.total_pagos    
    return data_unidad[["unidad_id","expensa_mes_pago","prop_impagos"]]
```

```python
get_cum_by_unidad("10")
```

```python
dfs = list()
for unidad in df.unidad_id.unique():
    data_unidad = get_cum_by_unidad(unidad)
    dfs.append(data_unidad)
```

```python
resumen_unidades = pd.concat(dfs)
```

```python
test_all_features[['expensa_id', 'consorcio_id', 'consorcio_nombre', 'unidad_id', 'unidad_denominacion',
       'unidad_tipo', 'expensa_proporcion', 'expensa_mes',
       'expensa_fecha', 'expensa_mes_pago', 'consorcio_impagos', 'consorcio_cantidad_expensas',
       'consorcio_cantidad_efectivo', 'consorcio_prop_impagos', 'consorcio_prop_efectivo',
       'unidad_prop_impagos','target', 'predicted']]
```

```python
results_by_consorcio = test_all_features[["consorcio_id","consorcio_nombre","target","predicted"]]
```

```python
results_by_consorcio["TP"] = test_all_features.target.map(bool) & test_all_features.predicted.map(bool)
results_by_consorcio["TN"] = ~test_all_features.target.map(bool) & ~test_all_features.predicted.map(bool)
results_by_consorcio["FP"] = ~test_all_features.target.map(bool) & test_all_features.predicted.map(bool)
results_by_consorcio["FN"] = test_all_features.target.map(bool) & ~test_all_features.predicted.map(bool)
```

```python
results_by_consorcio = results_by_consorcio.groupby(["consorcio_id","consorcio_nombre"]).sum()
```

```python
results_by_consorcio["Total"] = results_by_consorcio.TP + results_by_consorcio.TN + results_by_consorcio.FP + results_by_consorcio.FN
results_by_consorcio["PP"] = results_by_consorcio.TP + results_by_consorcio.FP
results_by_consorcio["PN"] = results_by_consorcio.TN + results_by_consorcio.FN
results_by_consorcio["RP"] = results_by_consorcio.TP + results_by_consorcio.FN
results_by_consorcio["RN"] = results_by_consorcio.FP + results_by_consorcio.TN
```

```python
results_by_consorcio["accuracy"] = (results_by_consorcio.TP + results_by_consorcio.TN)/results_by_consorcio.Total
results_by_consorcio["precision"] = (results_by_consorcio.TP)/results_by_consorcio.PP
results_by_consorcio["recall"] = results_by_consorcio.TP / results_by_consorcio.RP
```
