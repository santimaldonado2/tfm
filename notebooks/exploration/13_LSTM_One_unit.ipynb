```python
def build_timeseries(mat, n_clases, y_col_indexes):
    # y_col_index is the index of column that would act as output column
    # total number of time-series samples would be len(mat) - TIME_STEPS
    dim_0 = mat.shape[0] - TIME_STEPS
    dim_1 = mat.shape[1]
    x = np.zeros((dim_0, TIME_STEPS, dim_1))
    y = np.zeros((dim_0, n_clases))
    
    for i in range(dim_0):
        x[i] = mat[i:TIME_STEPS+i]
        for col_index in y_col_indexes:
            y[i, col_index - y_col_indexes[0]] = mat[TIME_STEPS+i, col_index]
    print("length of time-series i/o",x.shape,y.shape)
    return x, y
```

```python
X,y = build_timeseries(data_unidad.values, 3, [44,45,46])
```

```python
X_t = X[2:47,:,:]
y_t = y[2:47,:]
X_val = X[41:47,:,:]
y_val = y[41:47]
X_test = X[47:,:,:]
y_test = y[47:]
```

```python
BATCH_SIZE = 3
TIME_STEPS = 6
```

```python
from keras.layers.core import Dense, Dropout, Activation
from keras.models import Sequential
from keras.layers import LSTM

lstm_model = Sequential()
lstm_model.add(LSTM(100, batch_input_shape=(BATCH_SIZE, TIME_STEPS, X.shape[2]), dropout=0.0, recurrent_dropout=0.0, stateful=True, kernel_initializer='random_uniform'))
lstm_model.add(Dropout(0.5))
lstm_model.add(Dense(20,activation='relu'))
lstm_model.add(Dense(3,activation='sigmoid'))
lstm_model.compile(loss='categorical_crossentropy', optimizer="adam")
```

```python
lstm_model.fit(X_t, y_t,  validation_data=(X_val, y_val), epochs=10, verbose=2, batch_size=BATCH_SIZE,
                    shuffle=False)
```

```python
lstm_model.predict_classes(X_test,batch_size=3)
```
